```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('car_data.csv')
df
```

```python
df.duplicated(keep = 'first').sum()
```

```python
duplicate_indices = df[df.duplicated()].index
duplicate_indices
```

```python

average_city_mpg = df['city_mpg'].mean()
average_highway_mpg = df['highway_mpg'].mean()
average_combination_mpg = df['combination_mpg'].mean()

print(f'Average City MPG: {average_city_mpg:.2f}')
print(f'Average Highway MPG: {average_highway_mpg:.2f}')
print(f'Average Combined MPG: {average_combination_mpg:.2f}')

```

```python

fuel_type_distribution = df['fuel_type'].value_counts()
print(fuel_type_distribution)
```

```python

fuel_type_distribution.plot(kind = 'pie',autopct='%1.1f%%', figsize=(7, 7),colors=['lightblue', 'lightgreen', 'orange'])
plt.title('Distribution of Fuel Types')
plt.show()
```

```python
cylinders_fuel_efficiency = df.groupby('cylinders')[['city_mpg', 'highway_mpg','combination_mpg']].mean()
print(cylinders_fuel_efficiency)
```

```python
cylinders_fuel_efficiency.plot(kind= 'bar', figsize=(10, 6), color=['lightblue', 'lightgreen', 'orange'])
plt.title('Average Fuel Efficiency By Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('MPG')
plt.xticks(rotation = 0)
plt.show()

```

```python
drivetrain_efficiency = df.groupby('drive')['combination_mpg'].mean()

drivetrain_efficiency
```

```python
drivetrain_efficiency.plot(kind='bar', figsize=(8, 5), color='coral')
plt.title('Average Combined MPG by Drivetrain Type')
plt.xlabel('Drivetrain Type')
plt.ylabel('Average Combined MPG')
plt.xticks(rotation=0)
plt.show()
```

```python
electric_cars = df[df['fuel_type'] == 'electric']
gasoline_cars = df[df['fuel_type'] == 'gasoline']


avg_comb_mpg_electric = electric_cars['combination_mpg'].mean()
avg_comb_mpg_gasoline = gasoline_cars['combination_mpg'].mean()

print(f'Average Combined MPG for Electric Cars: {avg_comb_mpg_electric}')
print(f'Average Combined MPG for Gasoline Cars: {avg_comb_mpg_gasoline}')
```

```python
fuel_efficiency_trend = df.groupby('year')['combination_mpg'].mean()


print(fuel_efficiency_trend)

fuel_efficiency_trend.plot(kind='line', figsize=(10, 6), marker='o', color='blue')
plt.title('Average Fuel Efficiency Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Average Combined MPG')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
displacement_trend = df.groupby('year')['displacement'].mean()

print(displacement_trend)

displacement_trend.plot(kind='line', figsize=(10, 6), marker='o', color='green')
plt.title('Average Engine Displacement Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Average Displacement (L)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
cylinder_drivetrain_relationship = df.pivot_table(index='drive', columns='cylinders', aggfunc='size', fill_value=0)

print(cylinder_drivetrain_relationship)

cylinder_drivetrain_relationship.plot(kind='bar', figsize=(10, 6))
plt.title('Number of Cylinders by Drivetrain Type')
plt.xlabel('Drivetrain Type')
plt.ylabel('Count of Cars')
plt.xticks(rotation=45)
plt.legend(title='Number of Cylinders')
plt.grid(axis='y')
plt.show()
```
