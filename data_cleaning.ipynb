```python
#Here I imported pandas as the necessary library to carry out this assignment
import pandas as pd

#Here i defined the paths for the excel files to be loaded and the necessary operations to be done on them after
Mars_june = pd.read_excel(r"/home/murungi-nathan/Desktop/TakeHomeAssignment/Data/Mars/Mars_June_Data.xlsx")
Mars_july = pd.read_excel(r"/home/murungi-nathan/Desktop/TakeHomeAssignment/Data/Mars/Mars_July_Data.xlsx")
Mars_august = pd.read_excel(r"/home/murungi-nathan/Desktop/TakeHomeAssignment/Data/Mars/Mars_August_Data.xlsx")

Mercury_january = pd.read_excel(r"/home/murungi-nathan/Desktop/TakeHomeAssignment/Data/Mercury/Mercury_January_Data.xlsx")
Mercury_february = pd.read_excel(r"/home/murungi-nathan/Desktop/TakeHomeAssignment/Data/Mercury/Mercury_February_Data.xlsx")
Mercury_march = pd.read_excel(r"/home/murungi-nathan/Desktop/TakeHomeAssignment/Data/Mercury/Mercury_March_Data.xlsx")

Venus_january = pd.read_excel(r"/home/murungi-nathan/Desktop/TakeHomeAssignment/Data/Venus/Venus_January_Data.xlsx")
Venus_february = pd.read_excel(r"/home/murungi-nathan/Desktop/TakeHomeAssignment/Data/Venus/Venus_February_Data.xlsx")
Venus_march = pd.read_excel(r"/home/murungi-nathan/Desktop/TakeHomeAssignment/Data/Venus/Venus_March_Data.xlsx")


#Here I used this function to display all rows and columns so that i can easily inspect and observe the entire data sets
pd.set_option('display.max_rows', None)  
pd.set_option('display.max_columns', None) 
```

```python
# Here i defined a function and called it 'all_cooperatives' which concatenated, cleaned and aggregated the datafrom the multiple 
# monthly dataframes for three remaining cooperatives that is Mars,Mercury,and Venus. 
# This helped to remove duplicates,replace negative values with zero, and grouped the data by farmer names, summing their deliveries then 
# the cleaned dataframes were returned.

def all_cooperatives():
    combined_mars = pd.concat([Mars_june, Mars_july, Mars_august], axis=0, ignore_index=True).fillna(0)
    combined_mercury = pd.concat([Mercury_january, Mercury_february, Mercury_march], axis=0, ignore_index=True).fillna(0)
    combined_venus = pd.concat([Venus_january, Venus_february, Venus_march], axis=0, ignore_index=True).fillna(0)

    combined_mars = combined_mars.drop_duplicates()
    combined_mercury = combined_mercury.drop_duplicates()
    combined_venus = combined_venus.drop_duplicates()


    for df in [combined_mars, combined_mercury, combined_venus]:
        numeric_columns = df.select_dtypes(include=['number'])
        df[numeric_columns.columns] = numeric_columns.clip(lower=0)

    combined_mars = combined_mars.groupby(['name']).sum().reset_index()
    combined_mercury = combined_mercury.groupby(['name']).sum().reset_index()
    combined_venus = combined_venus.groupby(['name']).sum().reset_index()

    
    return combined_mars, combined_mercury, combined_venus


new_combined_mars, new_combined_mercury, new_combined_venus = all_cooperatives()

# Print the results
print(new_combined_mars)
print(new_combined_mercury)
print(new_combined_venus)

#Here i saved the cleaned dataframes to an excel file and named them accordingly
new_combined_mars.to_excel('combined_mars_cleaned.xlsx', index=False)
new_combined_mercury.to_excel('combined_mercury_cleaned.xlsx', index=False)
new_combined_venus.to_excel('combined_venus_cleaned.xlsx', index=False)
```
