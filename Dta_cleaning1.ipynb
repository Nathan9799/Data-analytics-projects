```python
#Here I imported pandas as the necessary library to carry out this assignment
import pandas as pd

#Here i defined the paths for the excel files to be loaded and the necessary operations to be done on them after
arrakis_jan = pd.read_excel(r"/home/murungi-nathan/Desktop/TakeHomeAssignment/Data/Arrakis/Arrakis_January_Data.xlsx")
arrakis_feb = pd.read_excel(r"/home/murungi-nathan/Desktop/TakeHomeAssignment/Data/Arrakis/Arrakis_February_Data.xlsx")
arrakis_march = pd.read_excel(r"/home/murungi-nathan/Desktop/TakeHomeAssignment/Data/Arrakis/Arrakis_March_Data.xlsx")

#Here I used this function to display all rows and columns so that i can easily inspect and observe the entire data set
pd.set_option('display.max_rows', None)  
pd.set_option('display.max_columns', None) 



```

```python
#Here I used the pd.concat() function to combine the the different excel files under the arrakis cooperative 
#and named the combined datasets combined_arrakis and i also used fillna(0) to replace the nan values with zeros

combined_arrakis = pd.concat([arrakis_jan, arrakis_feb, arrakis_march],axis = 0, ignore_index = True).fillna(0)
combined_arrakis
```

```python
#Here i used the info() function to get information about this dataframe to understand it better
combined_arrakis.info()
```

```python
#Here I wanted to find out the sum of duplicates which was 0
combined_arrakis.duplicated().sum()
```

```python
#Here i wanted to find the sum of null values 
combined_arrakis.isnull().sum()
```

```python
#I used the groupby() function here to group the data by the name of the farmers
#because the names would be repeated and i wanted to get a single record for each farmer and so by
#grouping by name allowed me to sum up the deliveries for each farmer

combined_arrakis_cleaned = combined_arrakis.groupby(['name']).sum().reset_index()
combined_arrakis_cleaned
```

```python
#Here i selected all the numeric columns from thr data frame and replaced the negative values with 0 to eliminate the negative values
numeric_columns = combined_arrakis_cleaned.select_dtypes(include=['number'])
numeric_columns[numeric_columns < 0] = 0


combined_arrakis_cleaned[numeric_columns.columns] = numeric_columns
combined_arrakis_cleaned
```

```python
#Here i saved the clean dataframe to an excel file and named it combined_arrakis_cleaned
combined_arrakis_cleaned.to_excel('combined_arrakis_cleaned.xlsx', index=False)
```
